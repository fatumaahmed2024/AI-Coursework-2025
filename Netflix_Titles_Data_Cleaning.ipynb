```python
#Q3
sa_movies_2021 = df[
    (df["country"] == "South Africa") &
    (df["release_year"] == 2021) &
    (df["type"] == "Movie")
]
print("🎬 South African Movies Released in 2021:\n")
for i, title in enumerate(sa_movies_2021["title"], start=1):
    print(f"{i}. {title}")
```

```python
#Q4
sa_movies_2021_cast = sa_movies_2021[["title", "cast"]]
print(sa_movies_2021_cast.to_string(index=False))
```
