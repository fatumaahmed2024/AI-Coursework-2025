```python
# Question 1: Convert days to seconds

# Ask user for input
days = int(input("Enter number of days: "))

# Calculate seconds- 1 day has 24 hrs, 1 hr has 60 mins, 1 min has 60 secs
seconds = days * 24 * 60 * 60

# Print result
print(f"{days} day(s) = {seconds} seconds")
```

```python
# Question 2: Volume of a sphere

#set pi
pi=3.142

# Ask user for radius
r = float(input("Enter the radius of the sphere: "))

# Compute volume
volume = (4/3) * pi * (r ** 3)

# Display result
print(f"Volume of sphere with radius {r} = {volume:.2f}") #2 decimal places
```

```python
# Question 3: Functions to compute area and perimeter of a square

# Define functions
def area(side):# there is one parameter called side
    return side * side

def perimeter(side):
    return 4 * side

# Ask user for side length
side_length = float(input("Enter the side length of the square: "))

# Call the functions and display both the results
print(f"Area = {area(side_length)}")
print(f"Perimeter = {perimeter(side_length)}")
```

```python
# Question 4: Determine whether a character input by user is uppercase or lowercase

def check_case(char):
    if char.isupper():
        return "Uppercase"
    elif char.islower():
        return "Lowercase"
    else:
        return "Not a letter"

# Ask user for a character
character = input("Enter a single character: ")

# ensure only 1 character is entered
if len(character) == 1:
    print(f"The character '{character}' is {check_case(character)}.")
else:
    print("Please enter one character only.")
```

```python
# Question 5: Converting pseudocode to Python

# Initializing variables
x = 0
y = 20

# Iterative loop
while y >= 6:
    y = y - 4  # subtract 4 from y
    x = x + (2 / y) # add 2/y to x

# Display result of x
print("Final value of x is :", x)
```

```python
# Question 6: Collect 5 values input by user and compute average

values = []  # empty array to store inputs

# Loop repeats 5 times
for i in range(5):
    num = float(input(f"Enter value {i+1}: "))
    values.append(num)

# Calculate average of values
average = sum(values) / len(values)

# Display results
print("Values entered:", values)
print("Average =", average)
```
